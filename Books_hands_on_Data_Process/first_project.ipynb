### Pandas function 


```python
### applying fucntion to series 

def MultiplyBy2(n):
    return n*2
adult_df.age.apply(MultiplyBy2)
```

```python
total_fnlwgt = adult_df.fnlwgt.sum()
def CalculatesPercentage(v):
    return v/total_fnlwgt*100

adult_df.fnlwgt = adult_df.fnlwgt.apply(CalculatesPercentage)
adult_df

## applyiing lambda function to find the same result
total_fnlwgt = adult_df.fnlwgt.sum()
adult_df.fnlwgt = adult_df.fnlwgt.apply(lambda v: v/total_fnlwgt*100)
adult_df
```

```python
# applying function to a dataframme
def CalcliFeNoED(row):
    return row.age - row['education-num']
adult_df.apply(CalcliFeNoED, axis=1)

#3 also by applying with lambda function
adult_df.apply(lambda r: r.age-r['education-num'], axis=1)
```

```python
# Applying a function – Analytic example 2 :: 
# Which one is more important in terms of your financial success: education or life experience? 
# using .corr() ---> pearson correlation

adult_df['lifeNoEd'] = adult_df.apply(lambda r: r.age-r['education-num'], axis=1) 
adult_df['capitalNet'] = adult_df.apply( lambda r: r.capitalGain - r.capitalLoss, axis=1)

adult_df[['education-num','lifeNoEd', 'capitalNet']].corr()
```
