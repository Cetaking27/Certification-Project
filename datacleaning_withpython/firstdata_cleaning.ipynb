### *data renaming axes indexes*

```python
data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                    index=['ohio', 'Colorado', 'New York'],
                    columns=['one', 'two', 'three', 'four'])
data
```

```python
transform = lambda x: x[:4].upper() # select first 4 rows
data.index.map(transform) # apply transform to index
# Index(['OHIO', 'COLO', 'NEW '], dtype='object')
data.index = data.index.map(transform) # apply transform to index
data
```

```python
data.rename(index=str.title, columns=str.upper) # apply transform to index and columns
```
