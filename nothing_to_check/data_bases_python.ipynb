#  Cours Data analyses avec python
Qu’est-ce qu’une donnée ? Signification et définition des données

Les données font référence à des collections de faits et d’informations individuelles.

Les données sont vitales pour la prise de décision, la planification et même pour raconter une histoire.

Il existe deux grands types de données générales :

Données qualitatives : sont des données exprimées en caractères non numériques. Elles sont des données exprimées en caractères non numériques.
Données quantitatives : sont exprimées en caractères numériques.

À la base, l’analyse des données consiste à identifier et à prédire les tendances, à déterminer les modèles, les corrélations et les relations dans les données disponibles, et à trouver des solutions à des problèmes complexes.

# importance de l'analyse des donnees:

- L’analyse des données améliore le ciblage des clients;
- L’analyse des données mesure le succès et la performance;
- L’analyse des données peut aider à résoudre les problèmes

# les etapes d'anayse de donnees:

- Étape 1 : Reconnaître et identifier les questions auxquelles il faut répondre (choisir l'objectif de l'analyse);
- Étape 2 : Collecte des données brutes : identifications du types de donnee (qualitative ou quantitative);
- Étape 3: Nettoyage des données (comprendre la structure et corriger les erreurs ainsi que filtrer ces données.) 
- Étape 4 : Analyse des données (analyser les données en fonction des questions et des objectifs de l’étape 1.)
  4 technique ce niveau:
  -) Analyse descriptive: « Que s’est-il passé ? la question a repondre
  -) Analyse diagnostique: Pourquoi cela s’est-il produit? ».
  -) Analyse prédictive: « Que risque-t-il de se passer à l’avenir ?
  -) Analyse prescriptive : « Que faut-il faire ? »
- Étape 5 : Partage des résultats: (consiste à interpréter vos conclusions.)

```python
'''
How to get or import data from website link using a code 

'''
def get_historical(symbol, exchange='bitfinex', after='201-03-01'):
    url ="https://api.cryptowat.ch/markets/{exchange}/{symbol}usd/ohlc".format(symbol=symbol, exchange=exchange)
    resp = requests.get(url, params={
        'periods':'3600',
        'after':str(int(pd.Timestamp(after).timestamp()))
                                     })
    resp.raise_for_status()
    data = resp.json()
    df = pd.DataFrame(data['result']['3600'], columns=['CloseTime','OpenPrice','HighPrice','LowPrice','ClosePrice','Volume','NA'])
    df['CloseTime'] = pd.to_datetime(df['CloseTime'], unit='s')
    df.set_index('CloseTime', inplace=True)
    return df 
```
